Êtes-vous fan de la série Game of Thrones ? Testons vos connaissances. Utilisons l'API : https://anapioficeandfire.com/ pour en savoir plus sur la série.

1. Importez les bibliothèques suivantes :

      * ```requests```
      * ```pandas```
      * ```json```

````python
import requests
import pandas as pd
import json
````

2. Accédez à l'URL indiquée dans la description ci-dessus. À l'aide de `requests`, obtenez des données sur un personnage de la série. Peu importe qui est le personnage.


````python
# Spécifiez l'URL
url = "https://anapioficeandfire.com/api/characters/583"

# Envoyez une requête GET à l'URL
response = requests.get(url)
````

3. Extrayez ces données au format JSON :

````python
# Convertissez la réponse en JSON
data = response.json()

# Affichez les données
print(json.dumps(data, indent=4))
````

4. Extrayez maintenant les clés JSON que vous avez obtenues et créez un DataFrame qui contiendra les clés JSON que vous avez extraites sous forme de colonne

**Astuce**: on peut utiliser: `r.json().keys()`.

````python
# Obtenez les clés
keys = data.keys()

# Créez un DataFrame à partir des clés
df = pd.DataFrame(columns=keys)

# Affichez le DataFrame
print(df)
````

5. Créez une boucle pour extraire les données des 150 premiers caractères de la série et ajoutez-les à votre DataFrame.

````python
# Créez un DataFrame vide avec les clés appropriées
url = "https://anapioficeandfire.com/api/characters/1"
response = requests.get(url)
data = response.json()
df = pd.DataFrame(columns=data.keys())

# Boucle sur les 150 premiers personnages
for i in range(1, 151):
    # Construisez l'URL pour chaque personnage
    url = f"https://anapioficeandfire.com/api/characters/{i}"

    # Faites la requête à l'API
    response = requests.get(url)

    # Si la requête est réussie, ajoutez les données au DataFrame
    if response.status_code == 200:
        data = response.json()
        df = df.append(data, ignore_index=True)

df
````

````python
# Créez un DataFrame vide avec les clés appropriées
url = "https://anapioficeandfire.com/api/characters/1"
response = requests.get(url)
data = response.json()
df = pd.DataFrame(columns=data.keys())

# Boucle sur les 150 premiers personnages
for i in range(1, 151):
    # Construisez l'URL pour chaque personnage
    url = f"https://anapioficeandfire.com/api/characters/{i}"

    # Faites la requête à l'API
    response = requests.get(url)

    # Si la requête est réussie, ajoutez les données au DataFrame
    if response.status_code == 200:
        data = response.json()
        df = pd.concat([df, pd.DataFrame([data])], ignore_index=True)

df
````
