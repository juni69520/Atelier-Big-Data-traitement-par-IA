3. Appelez l'API pour obtenir la liste des **250 premières crypto-monnaies** en termes de capitalisation boursière :
- Le résultat de la réponse doit être classé par capitalisation boursière dans l'ordre décroissant.
- Le paramètre **vs_currency** doit être défini sur USD
- Stocker les **10 premières crypto-monnaies** dans un nouvel objet python

```python
import requests

# Définir l'URL de l'API et les paramètres
url = "https://api.coingecko.com/api/v3/coins/markets"
params = {
    "vs_currency": "USD",
    "order": "market_cap_desc",
    "per_page": 250,
    "page": 1
}

# Appeler l'API
response = requests.get(url, params=params)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Extraire les données JSON
    cryptocurrencies = response.json()

    # Stocker les 10 premières crypto-monnaies dans un nouvel objet
    top_10_cryptocurrencies = cryptocurrencies[:10]

    # Afficher les 10 premières crypto-monnaies
    for crypto in top_10_cryptocurrencies:
        print(f"{crypto['name']} - Market Cap: ${crypto['market_cap']}")
else:
    print("Failed to retrieve data")
```

5. Obtenez les prix OHLC historiques du bitcoin sur depuis le début d'année :
* Le paramètre **vs_currency** doit être défini sur USD
* Le premier élément des listes de réponses (ex : 1632355200000) est appelé Unix Timestamp, qui est un moyen de suivre le temps sous la forme d'un total cumulé de secondes (cf. https://www.unixtimestamp.com/ ). Trouvez un moyen, en utilisant python, de convertir ce format en dates lisibles :
* Astuce 1 : vous devez supprimer les 3 derniers zéros de chaque élément d'horodatage unix pour une conversion correcte
* Astuce 2 : vous pouvez utiliser la bibliothèque datetime  pour procéder à la conversion (format de date attendu : année-mois-jour). Voici un exemple : [stackoverflow](https://stackoverflow.com/questions/3682748/converting-unix-timestamp-string-to-readable-date)

```python
import requests
from datetime import datetime

# Calculer le nombre de jours depuis le début de l'année
date1 = datetime(2023, 1, 1)
date2 = datetime.now()
difference = (date2 - date1).days

# Définir l'URL de l'API et les paramètres
url = "https://api.coingecko.com/api/v3/coins/bitcoin/ohlc"
params = {
    "vs_currency": "USD",
    "days": "365"
}

# Appeler l'API
response = requests.get(url, params=params)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Extraire les données JSON
    ohlc_data = response.json()

    # Parcourir les données OHLC et convertir les timestamps
    for data_point in ohlc_data:
        # Supprimer les 3 derniers zéros de l'horodatage Unix
        timestamp = data_point[0] // 1000
        # Convertir en date lisible
        date = datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d')
        # Afficher la date et les données OHLC
        print(f"Date: {date}, OHLC: {data_point[1:]}")
else:
    print("Failed to retrieve data")
```

En lisant la documentation du point de terminaison, je me suis rendu compte que je ne peux pas obtenir des enregistrements quotidiens car la granularité des données renvoyées par l'endpoint /coins/{id}/ohlc est automatiquement déterminée par CoinGecko et ne peut pas être ajustée. 6. Compte tenu des étapes précédentes, créez un dictionnaire contenant les prix OHLC historiques pour les dix premières cryptomonnaies sur la période en cours depuis le début de l'année


```python
import requests

# Obtenir la liste des 10 premières crypto-monnaies
url = "https://api.coingecko.com/api/v3/coins/markets"
params = {
    "vs_currency": "USD",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1
}
response = requests.get(url, params=params)
top_cryptos = response.json()

# Dictionnaire pour stocker les données OHLC
ohlc_dict = {}

# Pour chaque crypto-monnaie, obtenir les prix OHLC historiques
for crypto in top_cryptos:
    crypto_id = crypto['id']
    url = f"https://api.coingecko.com/api/v3/coins/{crypto_id}/ohlc?vs_currency=USD&days=180"
    response = requests.get(url)
    if response.status_code == 200:
        ohlc_data = response.json()
        ohlc_dict[crypto_id] = ohlc_data

# Afficher le dictionnaire
print(ohlc_dict)
```

7. Enregistrer l'historique dans un fichier .json grâce à la librairie json


```python
import json

# Enregistrer les données dans un fichier .json
with open('crypto_ohlc_history.json', 'w') as file:
    json.dump(ohlc_dict, file)
```
