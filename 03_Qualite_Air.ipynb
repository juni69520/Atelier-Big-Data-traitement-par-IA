6. Utilisez une requête pour répertorier tous les différents lieux en France, utilisez un paramètre pour pouvoir tous les obtenir en une seule requête !

11. Comme vous pouvez le voir, une requête couvrant 10 jours pour un seul emplacement contient déjà beaucoup de mesures dans ce cas. Imaginez si nous voudrions obtenir des données historiques pour les dix années précédentes pour l'ensemble du pays, cela prendrait du temps et beaucoup d'espace, alors pour l'instant, essayez simplement de créer une boucle pour obtenir 30 jours de données pour l'ensemble du pays.

```python
import requests
from datetime import datetime, timedelta

# URL de l'API
url = "https://api.openaq.org/v2/measurements"

# Paramètres de la requête
params = {
    "country": "FR",  # Code à deux lettres de la France
    "limit": 10000,  # Limite de résultats par page (valeur maximale)
    "parameter": "pm25",  # Paramètre de mesure de la qualité de l'air (pm25 pour les particules fines)
}

# Déterminer la plage de dates
end_date = datetime.now()
start_date = end_date - timedelta(days=30)

# Boucle pour obtenir les données par tranches de 30 jours
while start_date <= end_date:
    # Formater les dates au format AAAA-MM-JJTHH:MM:SS+HH:MM
    start_date_formatted = start_date.strftime('%Y-%m-%dT%H:%M:%S+00:00')
    end_date_formatted = (start_date + timedelta(days=30)).strftime('%Y-%m-%dT%H:%M:%S+00:00')

    # Mettre à jour les paramètres de la requête avec les dates
    params["date_from"] = start_date_formatted
    params["date_to"] = end_date_formatted

    # Effectuer la requête
    response = requests.get(url, params=params)

    # Vérifier si la requête a réussi
    if response.status_code == 200:
        # Extraire les données JSON
        data = response.json()

        # Vérifier si des mesures sont disponibles
        measurements = data['results']
        if measurements:
            # Afficher les données pour chaque emplacement
            for measurement in measurements:
                print(f"Emplacement : {measurement['location']}")
                print(f"Date : {measurement['date']['utc']}")
                print(f"Valeur : {measurement['value']} {measurement['unit']}")
                print("--------------------")

    else:
        print("Échec de la requête")

    # Passer à la prochaine tranche de 30 jours
    start_date += timedelta(days=30)
```
